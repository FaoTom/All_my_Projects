```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import special
```

```python
def firstpart(x, n, m, omega):
    return 1./np.sqrt(2**n*math.factorial(n))*((m*omega)/(np.pi))**0.25*np.exp(-(m*omega*x**2)/(2.))
def hermite(x, n):
    m=1
    omega=1
    p_monic = special.hermite(n)
    return (1./np.sqrt(2**n*math.factorial(n))*((m*omega)/(np.pi))**0.25*np.exp(-(m*omega*x**2)/(2.)))*p_monic(x)
```
