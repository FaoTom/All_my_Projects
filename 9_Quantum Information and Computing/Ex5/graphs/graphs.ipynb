```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy import special
```

```python
x = np.loadtxt('x.dat', dtype=float)
p = np.loadtxt('p.dat', dtype=float)
y = np.loadtxt('y.dat', dtype=float)
zx = np.loadtxt('zx.dat', dtype=float)
zp = np.loadtxt('zp.dat', dtype=float)
TT = 0.5
omega=3
pot = []
for tt in y:
    temp=0
    temp = [0.5*omega*(ii-tt/TT)**2 for ii in x]
    pot.append(temp)
```

```python

exp_x = np.loadtxt('expvalue.dat', dtype=float)
exp1sigma_x = np.loadtxt('expvalue1sigma.dat', dtype=float)

plt.imshow(zx[::-1,:], extent=(np.amin(x), np.amax(x), np.amin(y), np.amax(y)), aspect = 'auto',cmap='plasma', label='|\psi(x)|^2')
plt.colorbar()

contour = plt.contour(x,y,pot,100, alpha=0.5)
plt.plot(exp_x, y, color='gray',lw=2, ls='solid', label=r'$\langle x \rangle_{\psi_{GS}(x)}\pm\langle\Delta x\rangle_{\psi_{GS}(x)}$',alpha=0.8)
plt.plot(exp_x+exp1sigma_x, y,lw=2, ls='solid', color='gray', alpha=0.6)
plt.plot(exp_x-exp1sigma_x, y,lw=2 ,ls='solid', color='gray',alpha=0.6)
plt.xlabel('x', size=12)
plt.ylabel('Time', size=12)
plt.xlim(-10,10)
#plt.text(-2,2.5,r'$V_{pot}(t)$', color='white', size=14)
plt.legend(prop={'size': 12}, loc='upper left')
plt.title(r'Time evolution of $|\psi_{GS}(x)|^2$ $(\omega=1, T=0.5, m=5)$', size=12)
plt.savefig('evolution.pdf')
```
