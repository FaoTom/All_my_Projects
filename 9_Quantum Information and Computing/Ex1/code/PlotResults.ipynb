# Results plots

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cols=pd.read_csv('output_data/matmul_bycols_Ofast.txt',sep=",", header=None)
cols=cols.dropna(axis=1)

rows=pd.read_csv('output_data/matmul_byrows_Ofast.txt',sep=",", header=None)
rows=rows.dropna(axis=1)

matmul=pd.read_csv('output_data/matmul_fortran_Ofast.txt',sep=",", header=None)
matmul=matmul.dropna(axis=1)
```

```python
colsnof=pd.read_csv('output_data/matmul_bycols_noflag.txt',sep=",", header=None)
colsnof=colsnof.dropna(axis=1)

rowsnof=pd.read_csv('output_data/matmul_byrows_noflag.txt',sep=",", header=None)
rowsnof=rowsnof.dropna(axis=1)

matmulnof=pd.read_csv('output_data/matmul_fortran_noflag.txt',sep=",", header=None)
matmulnof=matmulnof.dropna(axis=1)
```

```python
cols00=pd.read_csv('output_data/matmul_bycols_O0.txt',sep=",", header=None)
cols00=cols00.dropna(axis=1)

rows00=pd.read_csv('output_data/matmul_byrows_O0.txt',sep=",", header=None)
rows00=rows00.dropna(axis=1)

matmul00=pd.read_csv('output_data/matmul_fortran_O0.txt',sep=",", header=None)
matmul00=matmul00.dropna(axis=1)
```

```python
cols01=pd.read_csv('output_data/matmul_bycols_O1.txt',sep=",", header=None)
cols01=cols01.dropna(axis=1)

rows01=pd.read_csv('output_data/matmul_byrows_O1.txt',sep=",", header=None)
rows01=rows01.dropna(axis=1)

matmul01=pd.read_csv('output_data/matmul_fortran_O1.txt',sep=",", header=None)
matmul01=matmul01.dropna(axis=1)
```

```python
cols02=pd.read_csv('output_data/matmul_bycols_O2.txt',sep=",", header=None)
cols02=cols02.dropna(axis=1)

rows02=pd.read_csv('output_data/matmul_byrows_O2.txt',sep=",", header=None)
rows02=rows02.dropna(axis=1)

matmul02=pd.read_csv('output_data/matmul_fortran_O2.txt',sep=",", header=None)
matmul02=matmul02.dropna(axis=1)
```

```python
cols03=pd.read_csv('output_data/matmul_bycols_O3.txt',sep=",", header=None)
cols03=cols03.dropna(axis=1)

rows03=pd.read_csv('output_data/matmul_byrows_O3.txt',sep=",", header=None)
rows03=rows03.dropna(axis=1)

matmul03=pd.read_csv('output_data/matmul_fortran_O3.txt',sep=",", header=None)
matmul03=matmul03.dropna(axis=1)
```

```python
plt.figure(figsize=(9,7))
plt.rcParams.update({'font.size': 16})
plt.plot(cols[0], rows[1], label='matmul_byrow')
plt.plot(cols[0], cols[1], label='matmul_bycol')

plt.plot(cols[0], matmul[1], label='MATMUL')


plt.yscale('log')
#plt.ylim(10e-5, 10)
plt.xlabel('$n_{col}=n_{row}$')
plt.ylabel('CPU Time [s]')
plt.grid(True, which="both", ls="-", color='0.9')
plt.legend()
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.title('matrix multiplication speed for different methods')
plt.savefig('ofast.png')
```

```python
plt.figure(figsize=(9,7))
plt.rcParams.update({'font.size': 16})
plt.plot(cols[0], cols[1], label='-Ofast')
plt.plot(colsnof[0], colsnof[1], label='No flags')
plt.plot(cols01[0], cols01[1], label='-O1')
plt.plot(cols02[0], cols02[1], label='-O2')
plt.plot(cols03[0], cols03[1], label='-O3')

plt.yscale('log')
plt.ylim(10e-5, 10)
plt.xlabel('$n_{col}=n_{row}$')
plt.ylabel('CPU Time [s]')
plt.grid(True, which="both", ls="-", color='0.9')
plt.legend()
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.title('matmul_bycol vs matrix size')
plt.savefig('matmulbycol.png')
```

```python
plt.figure(figsize=(9,7))
plt.rcParams.update({'font.size': 16})
plt.plot(matmul[0], matmul[1], label='-Ofast')
plt.plot(matmulnof[0], matmulnof[1], label='No flags')
plt.plot(matmul01[0], matmul01[1], label='-O1')
plt.plot(matmul02[0], matmul02[1], label='-O2')
plt.plot(matmul03[0], matmul03[1], label='-O3')

plt.yscale('log')
#plt.ylim(10e-5, 10)
plt.xlabel('$n_{col}=n_{row}$')
plt.ylabel('CPU Time [s]')
plt.grid(True, which="both", ls="-", color='0.9')
plt.legend()
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.title('MATMUL vs matrix size')
plt.savefig('matmulfortran.png')
```
