```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
gs2 = np.loadtxt('../data/gs2.txt')
gs3 = np.loadtxt('../data/gs3.txt')
gs4 = np.loadtxt('../data/gs4.txt')
```

```python
x = np.linspace(0.,3.,200)
xmin2 = x[x<2]
xmag2 = x[x>2]
plt.plot(x,gs2, label='$N=2$',lw=3)
plt.plot(x,gs3, label='$N=3$',lw=3)
plt.plot(x,gs4, label='$N=4$',lw=3)

plt.plot(xmin2, -1-xmin2**2/4,lw=3,color='orange')
plt.plot(xmag2, -xmag2,lw=3,label='MF',color='orange')
plt.axvline(x=1, color='black', ls='dashed', label=r'$\lambda=1$')
plt.xlabel(r'$\lambda$', size = 13)
plt.ylabel(r'$E_k/N$', size = 13)
plt.legend()
plt.xscale('log')
plt.grid()
plt.title('RSRG')
plt.savefig('RSRG.pdf')
```

```python
it2 = np.loadtxt('../data/it2.txt')
it3 = np.loadtxt('../data/it3.txt')
it4 = np.loadtxt('../data/it4.txt')
```
