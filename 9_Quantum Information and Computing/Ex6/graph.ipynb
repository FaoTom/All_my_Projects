```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
gen_2 = np.loadtxt('data/gen_2.txt')
gen_5 = np.loadtxt('data/gen_5.txt')
gen_6 = np.loadtxt('data/gen_6.txt')
gen_8 = np.loadtxt('data/gen_8.txt')

N_body_min = 2
N_body_max = 10
points  = 9
x = np.linspace(N_body_min, N_body_max, points)
plt.plot(x, gen_2, label='D=2', marker = 'd', lw=2)
plt.plot(x, gen_5, label='D=5', marker = 'd', lw=2)
plt.plot(x, gen_6, label='D=6', marker = 'd', lw=2)
plt.plot(x, gen_8, label='D=8', marker = 'd', lw=2)
plt.yscale('log')
plt.grid()
plt.legend()
plt.title('CPU Time vs system size N - General quantum state')
plt.ylabel('CPU TimeÂ [s]')
plt.xlabel('N')
plt.savefig('generic.pdf')
```

```python
from scipy.optimize import curve_fit

def func(x, a, b):    
    return a*x+b

sep_2 = np.loadtxt('data/sep_2.txt')
sep_3 = np.loadtxt('data/sep_3.txt')
sep_6 = np.loadtxt('data/sep_6.txt')
sep_5 = np.loadtxt('data/sep_5.txt')
sep_8 = np.loadtxt('data/sep_8.txt')


N_body_min = 10
N_body_max = 1000000
points  = 1000

x = np.arange(10,1000000,10000)

plt.plot(x, sep_3, label = 'D=3', marker='d', markersize=3)
plt.plot(x, sep_5, label = 'D=5', marker='d', markersize=3)
plt.plot(x, sep_6, label = 'D=6', marker='d', markersize=3)
plt.plot(x, sep_8, label = 'D=8', marker='d', markersize=3)

popt_x, pcov_x = curve_fit(func, x, sep_3)
results_x=func(x, popt_x[0], popt_x[1])
popt_x6, pcov_x6 = curve_fit(func, x, sep_6)
results_x6=func(x, popt_x6[0], popt_x6[1])
plt.plot(x,results_x, color='blue', label='$y=m_3+b_3$')
plt.plot(x,results_x6, color='green', label='$y=m_6+b_6$')
plt.grid()
plt.legend()
plt.title('CPU Time vs system size N - Separable quantum state')
plt.ylabel('CPU TimeÂ [s]')
plt.xlabel('System size N')
plt.savefig('separable.pdf')
```
